# 1.0 Análise Inicial dos Dados

O dataset Cross-Sectional (transversal) funciona como uma fotografia instantânea, capturando informações de 436 indivíduos em um único momento no tempo, o que permite comparar as diferenças cerebrais entre pessoas de diversas idades e grupos demográficos. Já o dataset Longitudinal é focado no acompanhamento progressivo, registrando 373 exames de idosos que retornaram para múltiplas visitas ao longo dos anos. Enquanto o primeiro é mais amplo e serve para identificar padrões gerais da anatomia cerebral, o segundo é fundamental para observar a velocidade da atrofia e a evolução da doença no mesmo paciente, contendo variáveis específicas como o número da visita (Visit) e o tempo de intervalo entre os exames (MR Delay).

### 1.4 Limpeza dos Dados

* Remoção de dados ausentes: Eliminar colunas que apresentem mais de 60% de valores faltantes (missing values).

* Limpeza de registros nulos: Remover todas as linhas onde a variável Group seja nula (NaN).

* Redução de dimensionalidade irrelevante: Descartar a coluna Hand, uma vez que ela possui apenas um valor único ("R") e não contribui para a diferenciação dos modelos.

* Codificação da variável alvo (Target): Criar uma nova coluna chamada Group, onde o valor 1 é atribuído a pacientes com Alzheimer (AD) e 0 para pacientes sem a doença.

* Imputação de dados: Criar uma instância do KNNImputer para preencher os valores faltantes restantes com base na similaridade dos vizinhos.

* Relatório de execução: Gerar e imprimir um relatório detalhando todas as alterações e imputações realizadas no dataset.

#### 1.4.1 Junção dos Datasets

```python
df_cross_sectional = df_cross_sectional.rename(columns={'ID': 'MRI ID', 'Educ': 'EDUC'}) 

df_merged = pd.concat([df_cross_sectional, df_logintudinal], axis=0, ignore_index=True)
df_merged
```

#### 1.4.2 Remoção de Missing Values

```python
def relatorio_valoresFaltantes(df):
    
    resumo_nulos = pd.DataFrame({
        'Coluna': df.columns,
        'Total de Nulos': df.isna().sum().values,
        'Porcentagem (%)': (df.isna().sum().values / len(df)) * 100
    })
    
    return resumo_nulos

def remover_coluna(df, colunas):
    
    df_filtrado = df.drop(columns=colunas)

    return df_filtrado
    
```

```python
relatorio_dadosFaltantes = relatorio_valoresFaltantes(df_merged)
relatorio_dadosFaltantes 
```

```python
limite = len(df_merged) * 0.4
df_merged = df_merged.dropna(thresh=limite, axis=1)
```

```python
df_merged
```

#### 1.4.3  Remoção de todas as linhas onde a variável Group seja nula (NaN).

```python
df_merged['Group'].isna().sum()
```

```python
df_merged = df_merged.dropna(subset=['Group'])
```

```python
df_merged['Group'].isna().sum()
```

#### 1.4.4 Redução de dimensionalidade irrelevante: Descartar a coluna Hand

```python
df_merged = remover_coluna(df_merged, ['Hand'])
```

```python
df_merged
```

#### 1.4.5 Codificação da variável alvo (Target)

```python
df_merged['Group'] = df_merged['Group'].apply(lambda x: 1 if x == 'Demented' else 0)
```

```python
df_merged
```

#### 1.4.6 Imputação de dados: Criar uma instância do KNNImputer para preencher os valores faltantes

```python
colunas_tabela3 = ['Age', 'CDR', 'SES', 'nWBV', 'Visit', 'MMSE', 'ASF', 'eTIV', 'Group', 'MR Delay']
df_final = df_merged[colunas_tabela3]

imputer = KNNImputer(n_neighbors=5) 
df_imputed = pd.DataFrame(imputer.fit_transform(df_final), columns=df_final.columns)
```

```python
df_imputed
```

```python
df_imputed.to_csv("dataset_tratado.csv")
```
