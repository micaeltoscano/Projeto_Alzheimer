
#### 1. Função de Custo: Log-Loss
Em Otimização Não-Linear, resolvemos este problema minimizando a **Negative Log-Likelihood** (ou Log-Loss). Esta função mede a discrepância entre as predições e os rótulos reais:

$$J(w) = -\frac{1}{N} \sum_{i=1}^{N} [y_i \log(h_w(x_i)) + (1 - y_i) \log(1 - h_w(x_i))]$$


```julia
using Convex

#Criando um exemplo de dados para regressão logística
N = 10                      # Número de exemplos
features = 3                # Número de variáveis (x)
X = randn(N, features)      # Matriz de dados X
y = rand(0:1, N)            # Vetor de respostas y (0 ou 1)
w = Variable(features)      # Variável de peso w
z = X * w                   # Produto escalar entre X e w

# Função de perda logística
J = sum(y[i] * logisticloss(-z[i]) + (1 - y[i]) * logisticloss(z[i]) for i in 1:N) * (1/N)

# Verificando a curvatura com a função custo da regressão logística
resultado_curvatura = vexity(J)

eh_convexa = (resultado_curvatura == Convex.ConvexVexity() || resultado_curvatura == Convex.AffineVexity())
# Imprimindo o resultado
println("A curvatura de J(w) é: ", eh_convexa ? "Convexa" : "Não Convexa")
```

#### 2. Função de Loss — Squared Hinge SVM

$$L(w) = \frac{1}{2}||w||^2 + C \sum_{i=1}^{N} \max(0, 1 - y_i \cdot x_i^T w)^2$$


```julia
N = 10                          # Número de exemplos
features = 3                    # Número de variáveis
X = randn(N, features)          # Matriz de dados X
y = rand([-1, 1], N)            # Vetor de respostas y (−1 ou 1)
w = Variable(features)          # Variável de peso w
z = X * w                       # Produto escalar entre X e w

C = 1.0                         # Constante de penalidade (hiperparâmetro da SVM)

reg = 0.5 * sumsquares(w)       # Termo de regularização (1/2 * ||w||^2)

loss = C * sum(square(pos(1 - y[i] * z[i])) for i in 1:N)  

# A função final L(w) é a soma das duas partes reg+loss
L = reg + loss

# Verificando a curvatura da função custo da SVM
resultado_curvatura = vexity(L)
eh_convexa = (resultado_curvatura == Convex.ConvexVexity() || resultado_curvatura == Convex.AffineVexity())

println("Curvatura da SVM: ", eh_convexa ? "Convexa" : "Não Convexa")
```

#### 2. Função de Loss — AdaBoost

$$
L = \sum_{i=1}^{n} \exp(-y_i F(x_i))
$$


```julia
n = 10                      # Número de exemplos
features = 3                # Número de variáveis
X = randn(n, features)      # Matriz de dados X
y = rand([-1, 1], n)        # Vetor de respostas y (−1 ou 1)

w = Variable(features)      # Variável de peso w

F_x = X * w                 # Produto escalar entre X e w

L = sum(exp(-y[i] * F_x[i]) for i in 1:n)

# Verificando a curvatura
resultado_curvatura = vexity(L)
eh_convexa = (resultado_curvatura == Convex.ConvexVexity() || resultado_curvatura == Convex.AffineVexity())

println("Curvatura da Perda do AdaBoost: ", eh_convexa ? "Convexa" : "Não Convexa")
```

Portanto, como todas as funções de custo são convexas, podemos aplicar os métodos de otimização não linear
